## **반복문 종료: break**

**반복문을 중간에 멈추고 싶다면, break를 활용하자!**  
* `break` : 반복문의 조건과 무관하게 반복문을 강제로 중단시켜줍니다. 
* 반복문으로 코드를 실행하던 중에 특정 조건에서 반복문을 종료하고 싶을 때 사용할 수 있습니다. 


#### **예제 6 : break 활용하기**

```python
#예제_6_break_활용하기 
while True :
    n = int(input("숫자를 입력하세요!"))
    if n % 3 ==0:
        print("3의 배수가 입력되었습니다! 프로그램을 종료합니다")
        break  # n%3==0 일 때 반복문을 종료합니다. 따라서 다음의 "n은 3의배수가 아닙니다!"은 출력되지 않습니다. 
    print("n은 3의배수가 아닙니다!")
```

## **처음부터 실행: continue**

**반복문을 중간까지만 실행하고 다시 처음으로 돌아오고 싶다면? continue를 활용하자!**

* continue 또한 break와 마찬가지로 반복문 내에서 사용됩니다.  
* `continue`는 반복문의 시작 부분으로 돌아가, for문의 변수를 진행하거나 while문의 반복조건을 판단하게 합니다.  
* 즉 반복 중에 continue를 만나면, continue 이후의 코드는 실행하지 않고, 다시 반복문의 조건 부분으로 돌아가서 조건이 부합하면 다음 반복을 이어서 수행합니다.   


#### **예제 7 : continue 활용하기**

```python
#예제_7_continue_활용하기 
while True :
    n = int(input("숫자를 입력하세요!"))
    if n % 3 ==0:
        print("3의 배수가 입력되었습니다! 프로그램을 종료합니다")
        break
    if n % 4 ==0:
        print("4의 배수가 입력되었습니다! 처음으로 돌아갑니다.")
        continue  # n%4==0 일 때 이후의 코드를 반복하지 않고, while문의 조건 부분으로 돌아갑니다. 
    print("n은 3의배수가 아닙니다!")
```
